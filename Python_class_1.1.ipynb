Q. A theme park allows entry if the person is either a VIP or has a ticket and is above 18.

```python

isVIP = input("Is the person VIP?\nYes or No\n")
if isVIP == "Yes" or isVIP == "yes":
  isVIP = True
else:
  isVIP = False

if isVIP:
  print("Allowed")

else:
  ticAvilable = input("Does the person have ticket?\nYes or No\n")
  if ticAvilable == "Yes" or ticAvilable== "yes":
    ticAvilable = True
  else:
    ticAvilable = False

  age = int(input("What is the age? "))
  if (age >= 18 and ticAvilable):
    print("Allowed")

  else:
    print("Not allowed")
```

Q.     Classify exam scores into:

    = 90 → Excellent
    70-89 → Good
    50-69 → Average
    <50 → Fail

Extension: Add special messages for borderline cases (e.g., 69 or 89).

```python
mark = int(input("Enter marks\n"))

if mark >= 90:
  print("Excellent")

elif mark>=70 and mark<=89:
  if mark==89:
    print("Almost to top teir")
  else:
    print("Good")

elif mark>=50 and mark<=69:
  if mark==69:
    print("Little more efford and you are better than this")
  else:
    print("Average")

elif mark<50:
  print("Fail")
```

for loop


In for loop we use **range(start,stop,step)**
1. default start is 0
2. stop value is excluded. ex. if 9 is the stop value, it will run till 8.
3. default step is 1

***Note:*** We can also write only stop, it will start from 0 and step by 1.

```python
for i in range(1,9,2):
  print(i)
```
